#**Question 1: What is hypothesis testing in statistics?**
Answer: Hypothesis Testing

Hypothesis testing is a statistical technique used to make inferences about a population parameter based on a sample of data. It involves formulating a hypothesis about the population parameter and then testing it using statistical methods.

Key Steps:

1. Null Hypothesis (H0): Formulate a null hypothesis that states there is no effect or no difference.
2. Alternative Hypothesis (H1): Formulate an alternative hypothesis that states there is an effect or a difference.
3. Test Statistic: Calculate a test statistic that measures the difference between the sample data and the null hypothesis.
4. P-Value: Calculate the p-value, which is the probability of observing the test statistic (or a more extreme value) assuming the null hypothesis is true.
5. Decision: Compare the p-value to a significance level (usually 0.05) and decide whether to reject the null hypothesis or not.

Applications:

1. Scientific Research: Hypothesis testing is widely used in scientific research to test hypotheses and make inferences about population parameters.
2. Quality Control: Hypothesis testing is used in quality control to determine whether a process is in control or not.
3. Business Decision-Making: Hypothesis testing is used in business decision-making to test hypotheses about customer behavior, market trends, and product performance.

#**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**
Answer: Null Hypothesis (H0)
The null hypothesis is a statement of no effect or no difference. It is a default assumption that there is no significant relationship or difference between variables. The null hypothesis is often denoted as H0.

**Key Differences:**

| Aspect        | Null Hypothesis (H₀)                          | Alternative Hypothesis (H₁ or Ha)              |
| ------------- | --------------------------------------------- | ---------------------------------------------- |
| Assumption    | No effect/difference                          | There is an effect/difference                  |
| Purpose       | To be tested and possibly rejected            | To be supported if H₀ is rejected              |
| Decision      | Retain or reject                              | Supported only if H₀ is rejected               |
| Risk of Error | Type I error occurs if H₀ is wrongly rejected | Type II error occurs if H₀ is wrongly retained |

#**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**
Answer: Significance Level (α)
The significance level, denoted by α (alpha), is a threshold probability value used in hypothesis testing to determine whether the observed data provides sufficient evidence to reject the null hypothesis. It represents the maximum probability of Type I error, which occurs when a true null hypothesis is incorrectly rejected.

Role in Hypothesis Testing
1. Determining the Critical Region: The significance level helps determine the critical region of the test statistic distribution. If the test statistic falls within the critical region, the null hypothesis is rejected.
2. Comparing with the P-Value: The significance level is compared with the p-value, which is the probability of observing the test statistic (or a more extreme value) assuming the null hypothesis is true. If the p-value is less than α, the null hypothesis is rejected.
3. Controlling Type I Error: The significance level controls the probability of Type I error. By setting α to a small value (e.g., 0.05), we can limit the probability of incorrectly rejecting a true null hypothesis.

Common Significance Levels
- α = 0.05: This is a commonly used significance level, which means there is a 5% chance of Type I error.
- α = 0.01: This significance level is more stringent, with a 1% chance of Type I error.
- α = 0.10: This significance level is less stringent, with a 10% chance of Type I error.

#**Question 4: What are Type I and Type II errors? Give examples of each.**
Answer: Type I Error
A Type I error occurs when a true null hypothesis is incorrectly rejected. This is often referred to as a "false positive" finding.

Type II Error
A Type II error occurs when a false null hypothesis is incorrectly retained. This is often referred to as a "false negative" finding.

Examples

Type I Error
- Medical Diagnosis: A person is diagnosed with a disease when they are actually healthy. In this case, the null hypothesis (the person is healthy) is incorrectly rejected.
- Court Trial: An innocent person is wrongly convicted of a crime. The null hypothesis (the person is innocent) is incorrectly rejected.

Type II Error
- Medical Diagnosis: A person with a disease is misdiagnosed as healthy. In this case, the null hypothesis (the person is healthy) is incorrectly retained.
- Quality Control: A batch of products is accepted as meeting quality standards when it actually does not. The null hypothesis (the batch meets quality standards) is incorrectly retained.

#**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**
Answer: **Z-test vs T-test: Key Differences**

| Feature                               | **Z-test**                       | **T-test**                                            |
| ------------------------------------- | -------------------------------- | ----------------------------------------------------- |
| **Population standard deviation (σ)** | Known                            | Unknown                                               |
| **Sample size (n)**                   | Large (typically n > 30)         | Small (typically n ≤ 30)                              |
| **Distribution used**                 | Standard Normal (Z) distribution | Student’s T distribution                              |
| **Shape of distribution**             | Fixed (bell curve)               | Varies depending on degrees of freedom (fatter tails) |
| **More sensitive to**                 | Large-sample precision           | Small-sample variability                              |

Choosing the Right Test

When deciding between a Z-test and a T-test, consider the sample size, population standard deviation, and research question. If the population standard deviation is known and the sample size is large, a Z-test may be suitable. However, if the population standard deviation is unknown or the sample size is small, a T-test is often more appropriate.

#**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**
**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

Answer:

Goal:
Perform a Z-test to determine whether the sample mean is significantly different from a known population mean.

Assumptions:
Population mean (μ₀): Let's assume it's 50.0

Population standard deviation (σ): Let's assume it's 0.5 (this is required for a Z-test)

Significance level (α): 0.05 (commonly used)

Hypotheses:
Null Hypothesis (H₀): μ = 50.0 → No significant difference.

Alternative Hypothesis (H₁): μ ≠ 50.0 → There is a significant difference.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population parameters
population_mean = 50.0
population_std = 0.5  # known σ

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
standard_error = population_std / np.sqrt(sample_size)

# Z-statistic
z_stat = (sample_mean - population_mean) / standard_error

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

**Interpreting the Output :**

Interpretation:

The sample mean (≈ 50.08) is slightly higher than the population mean (50.0).

The Z-statistic is about 1.99, and the p-value is 0.047.

Since p-value < 0.05, we reject the null hypothesis.

There is a statistically significant difference between the sample mean and the population mean at the 5% significance level.

#**Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mean = 100
std_dev = 15
sample_size = 100

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Step 2: Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level)/2)

margin_of_error = z_score * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Step 3: Plot histogram of the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')

plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Interpretation:

A Z-score of 0 = exactly at the mean.

Positive Z-score = above the mean.

Negative Z-score = below the mean.

Z = 2 → 2 standard deviations above the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Simulate some example data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first 5 Z-scores for reference
print("First 5 Z-scores:")
print(z_scores[:5])

# Plot the histogram of Z-scores
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
plt.legend()
plt.show()
```

**Explanation of Z-scores in the Plot:**

The histogram of Z-scores shows how the standardized values are distributed.

The red dashed line at Z = 0 shows the mean.

Most Z-scores typically fall between -2 and +2, especially in a normal distribution.

**Use Z-scores?**

They normalize data for comparison across different scales.

Help identify outliers (e.g., |Z| > 2 or 3).

Used in many statistical analyses (e.g., standardizing before clustering or regression).
